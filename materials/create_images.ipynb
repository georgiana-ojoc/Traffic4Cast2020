```python
import h5py
import imageio
import numpy as np
import os
```

```python
town = 'Berlin'  #@param ['Berlin', 'Istanbul', 'Moscow']

file_path = '/content/gdrive/My Drive/Licenta/Traffic4Cast/{}/files/{}_static_2019.h5'.format(town, town.upper())

images = '/content/gdrive/My Drive/Licenta/Traffic4Cast/{}/movies'.format(town)
```

```python
def get_data(file_path):
    file = h5py.File(file_path, 'r')
    group_key = list(file.keys())[0]
    data = np.array(file[group_key][:], dtype=np.float32)
    file.close()
    return data
```

```python
def create_image(data, images, file_path):
    frame = np.sum(data, axis=-1, dtype=np.uint8)
    path = os.path.join(images, os.path.basename(file_path)[:-2] + 'png')
    imageio.imwrite(path, frame)
    print('created image:', path)
```

```python
data = get_data(file_path)
create_image(data, images, file_path)
```
